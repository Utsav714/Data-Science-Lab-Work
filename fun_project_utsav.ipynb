```python
import time
import random

sentences = [
    "Data structures are fun",
    "Practice makes perfect",
    "Algorithms improve efficiency",
    "Coding is creativity",
    "Never stop learning"
]

leaderboard = []

def calculate_accuracy(original, typed):
    correct = sum(1 for o, t in zip(original, typed) if o == t)
    return (correct / len(original)) * 100

def typing_game():
    sentence = random.choice(sentences)
    print("\nType this:\n", sentence)

    input("Press Enter when ready...")

    start = time.time()
    typed = input("\nStart typing: ")
    end = time.time()

    time_taken = end - start
    wpm = (len(typed.split()) / time_taken) * 60
    accuracy = calculate_accuracy(sentence, typed)

    print(f"\nSpeed: {wpm:.2f} WPM")
    print(f"Accuracy: {accuracy:.2f}%")

    leaderboard.append(wpm)
    leaderboard.sort(reverse=True)

    print("\nTop Scores:")
    for i, score in enumerate(leaderboard[:5], 1):
        print(f"{i}. {score:.2f} WPM")

typing_game()
```
